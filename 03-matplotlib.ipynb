```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
```

```python
fig, axs = plt.subplots(2, 3, figsize=(15, 5))
```

```python
fig, axs = plt.subplots(2, 2, figsize=(15, 5))

axs[0,0].plot(np.random.randn(50).cumsum(), linestyle='-', color='b');
axs[0,1].plot(np.random.randn(50).cumsum(), linestyle='--', color='g');

```

```python
fig, axs = plt.subplots(2, 2, figsize=(15, 5))
axs[0,0].plot(np.random.randn(50).cumsum(), linestyle='--')
axs[0,1].hist(np.random.randn(100), bins=20, color='k', alpha=0.3)
axs[1,0].scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30));
fig.delaxes(axs[1, 1]) #  удаляем последний график
```

```python
plt.plot(np.random.randn(30).cumsum(), color='k', linestyle='dashed', marker='o');
```

```python
data = np.random.randn(30).cumsum()
plt.plot(data, linestyle='--', label='Default')
plt.plot(data, linestyle='-', drawstyle='steps-post', label='steps-post');
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 5))

ax.plot(np.random.randn(800).cumsum(), label='синий график')
ax.plot(np.random.randn(800).cumsum(), label='оранжевый график')

ax.set_xticks([0, 250, 500, 600, 750, 1000])
ax.set_xticklabels(['one', 'two', 'three', 'ffff', 'four', 'five'], rotation=45, fontsize='large')

ax.legend(loc='best')
ax.set_title('Мой график')
ax.set_xlabel('Шаги');
```
