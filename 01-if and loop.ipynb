## первый тип - цикл for

```python
for x in ['хлеб', 'яйца', 'вечина']:
    print(x)
```

```python
for x in ['хлеб', 'яйца', 'вечина']:
    print(x, end=' ')
```

```python
summa = 0
for x in [1, 2, 3, 4]:
    summa = summa + x
summa
```

```python
prod = 0
for x in [1, 2, 3, 4]:
    prod *= x
prod
```

```python
a = 'поищем буквы с разных словах ааааааа'
x = 'f'
for i in list(a):
    if x == i:
        print('буква есть')
        break # прервать цикл 'continue' вернуться в начало цикла
else:
    print('буквы нет')
```

вытащим данные из кортежа

```python
A = [(1, 2), (3, 4), (5, 6)]
for (a, b) in A:
    print(a, b)
```

вытащим данные из словаря

```python
A = {'a': 1, 'b': 2, 'c': 3 }
for key in A:
    print(key, ' -> ', A[key])
```

```python
list(A.items())
```

```python
for (key, value) in list(A.items()):
    print(key, ' -> ', value)
```

Спрятанный цикл в списках, множествах и т.п.

```python
message = 'грузите апельсины бочками'
vowels = {'в', 'а', 'л', 'ж', 'я'}

found = set()
for v in vowels:
    if v in message:
        found.add(v)
found    
```

```python
found2 = {v for v in vowels if v in message }
found2
```

## второй цикл - while

```python
x = 'spam'
while x:
    print(x, end = ' ')
    x = x[1:]
```

```python
a = 0; b = 10
while a < b:
    print(a, end=' ')
    a += 1
```

```python
y = 13
x = y // 2
while x > 1:
    if y % x == 0:
        print (y, " делится на ", x)
        break
    x -= 1
else:
    print(y, "простое число")
```
